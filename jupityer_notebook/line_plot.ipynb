# adding limits

```python
# first step always import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# 2nd step load data
flower=sns.load_dataset("iris")
flower

# 3rd step
# line plot always between 2 numeric value
sns.lineplot(x="sepal_length",y="sepal_width",data=flower)

# addong title
plt.title("flowers plot")

# adding limits
plt.xlim(2)
plt.ylim(1)
plt.show()
```
