```python
# first step always import libraries
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import median , mean

# 2nd step load data
ship=sns.load_dataset("titanic")
ship

# 3rd step
#  plot always between 2 numeric value
# sns.barplot(x="who",y="alone",hue="sex",data=ship)
sns.barplot(x="sex",y="fare",hue="who",data=ship,order=["female","male"],color="brown",
            estimator=mean,palette="pastel")
plt.show()
```
